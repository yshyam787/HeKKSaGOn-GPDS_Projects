```python
import nltk
from nltk.stem import WordNetLemmatizer, PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download necessary NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

documents = [
    "Those are cat food. They are for cats only.",
    "You should not eat those foods. It might be cat or dog food.",
    "Dogs do not eat cats. They eat dog food.",
    "Birds eat seeds, not cat food or dog food."
]

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()

def preprocess_text(text):
    tokens = word_tokenize(text)
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    tokens = [stemmer.stem(token) for token in tokens]
    tokens = [token for token in tokens if token not in stop_words]
    return ' '.join(tokens)

documents = [preprocess_text(doc) for doc in documents]
tfidf_vectorizer = TfidfVectorizer()

tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

print("Vocabulary: ", tfidf_vectorizer.vocabulary_)
print("TF-IDF Feature Matrix:\n", tfidf_matrix.toarray())
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from gensim.models import Word2Vec

# Data Preparation
train_documents = [
    "Cats eat cat food while dogs eat dog food.",
    "The birds might eat those seeds."
]

test_documents = [
    "Those are cat food. They are for cats only.",
    "You should not eat those foods. It might be cat or dog food.",
    "Dogs do not eat cats. They eat dog food.",
    "Birds eat seeds, not cat food or dog food."
]

# Preprocessing
def preprocess(document):
    lemmatizer = WordNetLemmatizer()
    stop_words = set(stopwords.words('english'))
    words = word_tokenize(document.lower())
    return [lemmatizer.lemmatize(word) for word in words if word.isalpha() and word not in stop_words]

# Preprocess the documents
processed_train_docs = [preprocess(doc) for doc in train_documents]
processed_test_docs = [preprocess(doc) for doc in test_documents]

# Training
model_cbow = Word2Vec(processed_train_docs, vector_size=100, window=4, min_count=1, sg=0)

# Generating Averaged Vectors
def average_vector(document, model):
    vectors = [model.wv[word] for word in document if word in model.wv]
    if len(vectors) > 0:
        return np.mean(vectors, axis=0)
    else:
        return np.zeros(model.vector_size)

averaged_docs_cbow = [average_vector(doc, model_cbow) for doc in processed_test_docs]

# Output the results
for doc, avg_vec in zip(test_documents, averaged_docs_cbow):
    print(f"Document: {doc}\nAveraged Vector: {avg_vec}\n")
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

# Data Preparation
train_documents = [
    "Cats eat cat food while dogs eat dog food.",
    "The birds might eat those seeds."
]

test_documents = [
    "Those are cat food. They are for cats only.",
    "You should not eat those foods. It might be cat or dog food.",
    "Dogs do not eat cats. They eat dog food.",
    "Birds eat seeds, not cat food or dog food."
]

# Preprocessing
def preprocess(document):
    lemmatizer = WordNetLemmatizer()
    stop_words = set(stopwords.words('english'))
    words = word_tokenize(document.lower())
    return [lemmatizer.lemmatize(word) for word in words if word.isalpha() and word not in stop_words]

# Preprocess the documents
processed_train_docs = [preprocess(doc) for doc in train_documents]
processed_test_docs = [preprocess(doc) for doc in test_documents]

# Training
model_skipgram = Word2Vec(processed_train_docs, vector_size=100, window=4, min_count=1, sg=1)

# Generating Averaged Vectors
def average_vector(document, model):
    vectors = [model.wv[word] for word in document if word in model.wv]
    if len(vectors) > 0:
        return np.mean(vectors, axis=0)
    else:
        return np.zeros(model.vector_size)

# Averaged vectors for documents - Skip-gram
averaged_docs_skipgram = [average_vector(doc, model_skipgram) for doc in processed_test_docs]

# Output the results
for doc, avg_vec in zip(test_documents, averaged_docs_skipgram):
    print(f"Document: {doc}\nAveraged Vector: {avg_vec}\n")
```

```python
import gensim
from gensim import corpora
from nltk.tokenize import word_tokenize

# Sample documents
documents = ["Data science is an evolving field",
             "The AI revolution is led by deep learning",
             "Cats are great pets",
             "Dogs are very loyal"]

# Tokenization
tokenized_docs = [word_tokenize(doc.lower()) for doc in documents]

# Creating a dictionary (ID-Word mapping)
dictionary = corpora.Dictionary(tokenized_docs)

# Creating a corpus (Document-Term matrix)
doc_term_matrix = [dictionary.doc2bow(doc) for doc in tokenized_docs]

# Creating the LDA model
LDA = gensim.models.ldamodel.LdaModel

lda_model = LDA(corpus=doc_term_matrix, id2word=dictionary, num_topics=2, random_state=100,
                chunksize=1000, passes=50)

# Print topics
for i, topic in lda_model.print_topics(-1):
    print(f"Topic: {i}, \nWords: {topic}\n")
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.decomposition import TruncatedSVD
import re
from umap import UMAP
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering

# Set random seed for reproducibility
np.random.seed(42)

# Download necessary NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Read the CSV file into a pandas DataFrame
train_url = "https://www.dropbox.com/scl/fi/j5fkr7gf9vqah539ktj2o/train_science.csv?rlkey=i6kqc5t2l9fuewtrngxyeddoa&st=jz1qnasy&dl=0&raw=1"
df_train = pd.read_csv(train_url, index_col=0)
df_train = df_train[~df_train['Comment'].str.contains('removed')]

# Preprocessing function
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    tokens = word_tokenize(text)
    tokens = [token for token in tokens if token not in stop_words]
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return ' '.join(tokens)

# Apply preprocessing to 'Comment' column
df_train['ProcessedComment'] = df_train['Comment'].apply(preprocess_text)

# TF-IDF Vectorization
vectorizer = TfidfVectorizer(max_features=1000, max_df=0.8, min_df=5)
X = vectorizer.fit_transform(df_train['ProcessedComment'])

# Dimensionality reduction with Truncated SVD
svd = TruncatedSVD(n_components=100, random_state=42)
X_processed = svd.fit_transform(X)

# Perform dimensionality reduction for visualization with UMAP
umap_model = UMAP(n_neighbors=100, min_dist=0.75, spread=2, random_state=42, n_jobs=1)
X_umap = umap_model.fit_transform(X_processed)

# Clustering (K-means)
num_clusters = 9  # Adjust the number of clusters as needed
agg_clustering = AgglomerativeClustering(n_clusters=num_clusters)
df_train['Cluster'] = agg_clustering.fit_predict(X_umap)

# Visualize the clusters
plt.figure(figsize=(8, 6))
for i in range(num_clusters):
    plt.scatter(X_umap[df_train['Cluster'] == i, 0], X_umap[df_train['Cluster'] == i, 1],
                label=f'Cluster {i}', s=50, alpha=0.7)
plt.title('UMAP Visualization of Clusters')
plt.xlabel('UMAP Component 1')
plt.ylabel('UMAP Component 2')
plt.legend()
plt.show()
```

```python
# Initialize stopwords, stemmer, and lemmatizer
import re
import string
from nltk.stem import SnowballStemmer
# stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
stemmer = SnowballStemmer('english')

def preprocess_text(text):
    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Convert to lowercase
    text = text.lower()

    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # Remove non-ASCII characters
    text = re.sub(r'[^\x00-\x7F]+', ' ', text)

    # Remove extra whitespaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenize
    tokens = word_tokenize(text)

    # Remove stopwords
    tokens = [token for token in tokens if token not in stop_words]

    # # Lemmatize
    # tokens = [lemmatizer.lemmatize(token) for token in tokens]

    # Stem
    tokens = [stemmer.stem(token) for token in tokens]

    # Generate bigrams
    # bigrams = list(ngrams(tokens, 2))
    # bigram_tokens = ['_'.join(bigram) for bigram in bigrams]
    # tokens = tokens + bigram_tokens

    # Join tokens back into a string
    processed_text = ' '.join(tokens)

    return processed_text

X_processed = X_train.apply(preprocess_text)

print("Before Preprocessing:")
print(X_train.tail())
print("After Preprocessing:")
print(X_processed.tail())
```
