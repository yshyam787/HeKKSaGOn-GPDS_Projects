```python
import pandas as pd
from scipy.stats import zscore

# Collect all numerical columns in the DataFrame
numerical_columns = df.select_dtypes(include=['number']).columns

# Calculate z-scores for each numerical column
z_scores_df = df[numerical_columns].apply(zscore)

# Identify outliers based on the threshold
outliers_mask = abs(z_scores_df) > 1.65

# Display outliers with original values
outliers_original_values = df[outliers_mask.any(axis=1)]

print("Outliers in numerical columns:")
print(outliers_original_values)
```

### **<font color="#819A00"> ➎ Train the Model**
</div>

<img src="https://i.imgur.com/puPTrbs.png" width=700 />
<img src="https://i.imgur.com/eVE7DKz.png" width=700 />


```python
# Read the CSV file
df = pd.read_csv("https://www.dropbox.com/scl/fi/zd2bv0elqg0io41k3iexk/golf_dataset_min.csv?rlkey=dxb9j1h6blay8s7zvj7rud00s&dl=0&raw=1")

# Map 'No' to False and 'Yes' to True in the 'Play' column
df['Play'] = df['Play'].map({'No': False, 'Yes': True})

# Convert categorical variables into dummy variables
df = pd.get_dummies(df).astype(int)

# Split the data into features (X) and target (y)
X = df.drop(columns=['Play','Wind'])
y = df['Play']
```

<img src="https://i.imgur.com/Er0NjDN.png" width=700 />


```python
# Split the data into training and testing sets
X_train, X_test = X[:-4], X[-4:]
y_train, y_test = y[:-4], y[-4:]
```

<img src="https://i.imgur.com/RK0j40w.png" width=700 />

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree

# Initialize the Decision Tree classifier
clf = DecisionTreeClassifier(random_state=42)

# Train the Decision Tree classifier on the training data
clf.fit(X_train, y_train)

# Convert class names to strings
class_names_str = [str(class_name) for class_name in clf.classes_]

# Plot the Decision Tree
plt.figure(figsize=(10,8))
plot_tree(clf, feature_names=X.columns, class_names=class_names_str, filled=True)
plt.title("Decision Tree Visualization")
plt.show()
```

### **<font color="#819A00"> ➏ Evaluate & Interpret the Model**
</div>

<img src="https://i.imgur.com/Fh24ign.png" width=700 />
<img src="https://i.imgur.com/rDJnZ82.png" width=700 />


```python
from sklearn.metrics import accuracy_score
```

<img src="https://i.imgur.com/zDmXNtg.png" width=700 />


```python
# Predict the target values on the testing data
y_pred = clf.predict(X_test)

# Create a DataFrame to compare actual and predicted values
comparison_df = pd.DataFrame({'Actual': y_test.reset_index(drop=True), 'Predicted': y_pred})

# Map numerical predictions to 'No' and 'Yes'
y_test_pred_mapped = comparison_df.replace({0: 'No', 1: 'Yes'})

# Show the DataFrame
print(y_test_pred_mapped)
```

<img src="https://i.imgur.com/mfAzaKm.png" width=700 />


```python
# Calculate the accuracy of the classifier
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
```

<img src="https://i.imgur.com/fV7VrwZ.png" width=700 />

```python
def extract_rules(tree, feature_names, class_names):
    rules = []

    def recurse(node, rule):
        if tree.feature[node] != -2:
            name = feature_names[tree.feature[node]]
            threshold = tree.threshold[node]
            left_rule = f"{name} <= {threshold}"
            right_rule = f"{name} > {threshold}"
            recurse(tree.children_left[node], rule + [left_rule])
            recurse(tree.children_right[node], rule + [right_rule])
        else:
            class_index = tree.value[node].argmax()
            class_name = class_names[class_index]
            rule_str = " and ".join(rule)
            rules.append(f"If {rule_str} then Play is {class_name}")

    recurse(0, [])
    return rules

rules = extract_rules(clf.tree_, X.columns, clf.classes_)
print("Rules Learned by the Model:")
for rule in rules:
    print(rule)
```

<img src="https://i.imgur.com/MCPBHxg.png" width=700 />

```python
from sklearn.model_selection import cross_val_score, KFold
# Perform 3-fold cross-validation
cv = KFold(n_splits=3, shuffle=False)
accuracies = cross_val_score(clf, X, y, cv=cv)

# Display the accuracy for each fold
for i, accuracy in enumerate(accuracies, 1):
    print(f"Accuracy for Fold {i}: {accuracy}")

# Average accuracy across all folds
print(f"Average Accuracy: {accuracies.mean().round(2)*100:.2f}%")
```

<img src="https://i.imgur.com/XPitxfk.png" width=700 />

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Define the models
models = [
    DecisionTreeClassifier(criterion='gini', max_depth=4, min_samples_split=2, random_state=0),
    DecisionTreeClassifier(criterion='entropy', max_depth=3, min_samples_split=4, random_state=0),
    KNeighborsClassifier(n_neighbors=3)
]

# Train and evaluate each model
for i, model in enumerate(models, 1):
    # Train the model
    model.fit(X_train, y_train)

    # Predict the target values on the testing data
    y_pred = model.predict(X_test)

    # Calculate the accuracy of the model
    accuracy = accuracy_score(y_test, y_pred)

    # Print the accuracy
    print(f"Model {i}: Accuracy = {accuracy:.2f}")
```
